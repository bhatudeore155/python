### Question 4:

We have a list of numbers given below. Print the square of these numbers into another list using list comprehension.

num = [2, 4, 6, 8]

#### Expected Output:

[4, 16, 36, 64]

```python
num = [2, 4, 6, 8]
num2 = []
for i in num:
    r = i*i
    num2.append(r)
print(num2)
```

```python
print([i**2 for i in num])
```

### Question 9:

Reverse the integer given below.

n = 5623

#### Expected Output:

3265

```python
n = 5623
reversed_num = int(''.join(reversed(str(n))))
print(reversed_num)
```

```python
| Method       | Code                             | Converts to String? | Output |
| ------------ | -------------------------------- | ------------------- | ------ |
| Slicing      | `int(str(n)[::-1])`              | ✅                   | 3265   |
| While Loop   | use `%` and `//`                 | ❌                   | 3265   |
| Recursion    | custom function                  | ❌                   | 3265   |
| `reversed()` | `int(''.join(reversed(str(n))))` | ✅                   | 3265   |
```

### Question 10:

We have a list of numbers given below. Print the number, square of the number, and the cube of the number, all in a single list.

num = [2, 4, 6, 8]

[(2, 4, 8), (4, 16, 64), (6, 36, 216), (8, 64, 512)]
